```python
train['GrLivArea'].min()
```

```python
# parametros de la recta
w = 50
b = 200000
```

```python
# puntos de la recta
x = np.linspace(0,train['GrLivArea'].max(),100)
y = w*x+b

# grafica de la recta
train.plot.scatter(x='GrLivArea',y='SalePrice')
plt.plot(x, y, '-r')
plt.ylim(0,train['SalePrice'].max()*1.1)
# plt.grid()
plt.show()
```
